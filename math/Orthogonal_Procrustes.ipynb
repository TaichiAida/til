# Orthogonal Procrustes
* optimize `min( base@R - other )` , and compute a rotate vector **R**.
* In Hamilton et al., R obtained above was **multiplied by other instead of base. (other @ R)**

â†’Is it really true?

```
import numpy as np
from scipy.linalg import orthogonal_procrustes
import matplotlib.pyplot as plt
```

```
np.random.seed(0)
N = 100
```

```
base = np.random.randn(N, 2)
base[:, 0] -= 10; base[:, 1] -= 10
other = np.random.randn(N, 2)
other[:, 0] += 10; other[:, 1] += 10
```

```
# normalize
base_norm = np.linalg.norm(base)
other_norm = np.linalg.norm(other)
base /= base_norm
other /= other_norm
```

```
plt.scatter(base[:, 0], base[:, 1], c='blue', label='base')
plt.scatter(other[:, 0], other[:, 1], c='red', label='other')
plt.legend(loc='upper left')
plt.show()
```

# min(base@R - other)

```
R_b2o, _ = orthogonal_procrustes(base, other)
```

```
plt.scatter(base@R_b2o[:, 0], base@R_b2o[:, 1], c='blue', label='base_fixed')
plt.scatter(other[:, 0], other[:, 1], c='red', label='other')
plt.legend(loc='upper left')
plt.show()
```

```
plt.scatter(base[:, 0], base[:, 1], c='blue', label='base')
plt.scatter(other@R_b2o[:, 0], other@R_b2o[:, 1], c='red', label='other_fixed')
plt.legend(loc='upper left')
plt.show()
```

# min(other@R - base)

```
R_o2b, _ = orthogonal_procrustes(other, base)
```

```
plt.scatter(base[:, 0], base[:, 1], c='blue', label='base')
plt.scatter(other@R_o2b[:, 0], other@R_o2b[:, 1], c='red', label='other_fixed')
plt.legend(loc='upper left')
plt.show()
```

```
plt.scatter(base@R_o2b[:, 0], base@R_o2b[:, 1], c='blue', label='base_fixed')
plt.scatter(other[:, 0], other[:, 1], c='red', label='other')
plt.legend(loc='upper left')
plt.show()
```

# Conclusion
both base@R and other@R can align.
